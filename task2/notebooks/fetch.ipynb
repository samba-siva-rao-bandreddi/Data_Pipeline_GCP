```python
import requests
import logging
import json

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def fetch_weather(latitude, longitude, hourly_fields):
    url = "https://api.open-meteo.com/v1/forecast"

    params = {
        "latitude": latitude,
        "longitude": longitude,
        "hourly": ",".join(hourly_fields)
    }

    try:
        logging.info("Fetching weather data...")

        response = requests.get(
            url=url,
            params=params,
            timeout=10
        )

        response.raise_for_status()

        data = response.json()

        logging.info("Weather data fetched successfully")

        return data

    except requests.exceptions.Timeout:
        logging.error("Request timed out")

    except requests.exceptions.HTTPError as e:
        logging.error(f"HTTP error occurred: {e}")

    except requests.exceptions.RequestException as e:
        logging.error(f"Request failed: {e}")

    return None

if __name__ == "__main__":

    latitude = 13.0827      # Chennai
    longitude = 80.2707

    hourly_fields = [
        "temperature_2m",
        "relative_humidity_2m",
        "wind_speed_10m"
    ]

    weather_data = fetch_weather(
        latitude,
        longitude,
        hourly_fields
    )

    if weather_data:
        with open("../data/raw/chennai.json","w") as f:
            json.dump(weather_data,f,indent=4)
        print("Weather data saved to chennai.json")
```

```python
cities = [
    {
        "city": "Chennai",
        "latitude": 13.0827,
        "longitude": 80.2707
    },
    {
        "city": "Hyderabad",
        "latitude": 17.3850,
        "longitude": 78.4867
    },
    {
        "city": "Bengaluru",
        "latitude": 12.9716,
        "longitude": 77.5946
    },
    {
        "city": "Mumbai",
        "latitude": 19.0760,
        "longitude": 72.8777
    },
    {
        "city": "Kolkata",
        "latitude": 22.5726,
        "longitude": 88.3639
    }
]
```
